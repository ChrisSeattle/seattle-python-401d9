# Regressions

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import numpy.polynomial.polynomial as poly
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = [12.0, 8.0]
```

```python
# Generate some random data, and split into training and test
def generate_data(size):
    X = np.random.rand(size) * 100.0
    Y = X ** 2 + 7
    Y = Y + np.random.randn(Y.shape[0]) * 300
    return X, Y
```

```python
X_full, Y_full = generate_data(20)

X_train, X_test, Y_train, Y_test = train_test_split(X_full, Y_full, test_size=0.5, random_state=42)
```

```python
# Plot the training data
plt.plot(X_train, Y_train, 'bo', markersize=20)
plt.title("Training Set")
```

```python
# Plot the test data
plt.plot(X_test, Y_test, 'bo', markersize=20)
plt.title("Test Set")
```

```python
# Fit different regressions to the plot
```

```python
coefs_lin = poly.polyfit(X_train, Y_train, 1)
coefs_sq = poly.polyfit(X_train, Y_train, 2)
coefs_ten = poly.polyfit(X_train, Y_train, 11)
```

```python
# Compute set of Y's
X_line = np.linspace(0, 100, 10000)

ffit_lin = poly.polyval(X_line, coefs_lin)
ffit_sq = poly.polyval(X_line, coefs_sq)
ffit_ten = poly.polyval(X_line, coefs_ten)
```

```python
plt.ylim(min(Y_train - 5000), max(Y_train + 5000))

plt.plot(X_line, ffit_lin, 'g')
plt.plot(X_line, ffit_sq, 'b')
plt.plot(X_line, ffit_ten, 'r')

plt.plot(X_train, Y_train, 'r.', markersize=20)
```
